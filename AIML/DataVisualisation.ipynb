```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.DataFrame({
    'Product': ['Laptop', 'Phone', 'Tablet', 'Watch', 'Headphones'],
    'Sales': [2500, 4200, 1800, 1200, 2800]
})
```

```python
np.random.seed(42)
test_scores = np.random.normal(70, 15, 200)
```

```python
np.random.seed(42)
n_points = 50
study_hours = np.random.uniform(1, 8, n_points)
exam_scores = 60 + 5 * study_hours + np.random.normal(0, 10, n_points)
scatter_data = pd.DataFrame({
    'Study_Hours': study_hours,
    'Exam_Score': exam_scores
})
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [5, 7, 12, 15, 20, 25, 28, 27, 22, 15, 10, 6]
line_data = pd.DataFrame({
    'Month': months,
    'Temperature': temperature
})
```

```python
np.random.seed(42)
correlation_matrix = pd.DataFrame(
    np.random.rand(5, 5),
    columns=['A', 'B', 'C', 'D', 'E'],
    index=['A', 'B', 'C', 'D', 'E']
)
```

```python
pie_data = pd.DataFrame({
    'Company': ['Company A', 'Company B', 'Company C', 'Company D'],
    'Market_Share': [30, 25, 20, 25]
})
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sales_data['Product'], sales_data['Sales'], color='skyblue')
plt.title('Product Sales Comparison')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(test_scores, bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Test Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(scatter_data['Study_Hours'], scatter_data['Exam_Score'], 
           alpha=0.5, color='purple')
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(line_data['Month'], line_data['Temperature'], 
         marker='o', linewidth=2, color='red')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', 
            vmin=0, vmax=1, center=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(pie_data['Market_Share'], labels=pie_data['Company'], 
        autopct='%1.1f%%', startangle=90)
plt.title('Market Share Distribution')
plt.axis('equal')
plt.show()
```
